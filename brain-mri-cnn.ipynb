```python
# 17 min to run 20 epochs 5x
import numpy as np
import pandas as pd
import cv2
import os
import warnings
import matplotlib.pyplot as plt
from matplotlib import gridspec

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from sklearn.model_selection import StratifiedKFold

# Set Matplotlib defaults
plt.rc('figure', autolayout=True)
plt.rc('axes', labelweight='bold', labelsize='large',
       titleweight='bold', titlesize=18, titlepad=10)
plt.rc('image', cmap='magma')
warnings.filterwarnings("ignore")  # to clean up output cells

# Read the data from ./input/brain-tumor-mri-dataset/combined
def read_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # makes it grey scale
    img = cv2.resize(img, (32, 32)) # resize to 32x32 pixels for efficiency
    return img

x = [] # store pictures
y = [] # store the labels of the pictures
label_names = {}
label_num = 0
for label in os.listdir("../input/combined/combined"):
    # use ordinal encoding
    label_names[label_num] = label
    for file in os.listdir(f"../input/combined/combined/{label}"):
        processed = read_image(f"../input/combined/combined/{label}/{file}")
        x.append(processed)
        y.append(label_num)
    label_num += 1

x = np.array(x)
y = np.array(y)

# Handling the data
class TumorDataset(Dataset):
    def __init__(self, images, labels):
        self.images = images
        self.labels = labels

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx].astype(np.float32) / 255.0
        label = self.labels[idx]
        return torch.tensor(image).unsqueeze(0), torch.tensor(label)

# Define the CNN model
class TumorCNN(nn.Module):
    def __init__(self):
        super(TumorCNN, self).__init__()
        self.layers = nn.Sequential(
            # Block 1
            
            # Normalizes input to network
            # Parameter indicates number of channels in input data: since it is greyscale, there is only 1
            nn.BatchNorm2d(1),
            # Extracts features from images
            # Input has 1 channel (grey scale), outputs 24 feature maps (channels)
            nn.Conv2d(1, 24, kernel_size=3, padding='same'),
            # Activation function
            nn.ReLU(),
            # Condenses the feature maps
            # kernel size 2x2 (this is TensorFlow's default)
            nn.Conv2d(24, 40, kernel_size=3, padding='same'),
            nn.ReLU(),
            nn.MaxPool2d(2),

            # Block 2
            nn.BatchNorm2d(40),
            nn.Conv2d(40, 80, kernel_size=3, padding='same'),
            nn.ReLU(),
            nn.Conv2d(80, 112, kernel_size=3, padding='same'),
            nn.ReLU(),
            nn.Conv2d(112, 192, kernel_size=3, padding='same'),
            nn.ReLU(),
            nn.MaxPool2d(2),

            # Conert 2D feature maps to 1D vectors
            nn.Flatten(),
            # Fully connected (aka dense) layers
            # Case 3 - (4 layers)
            nn.Linear(192 * 8 * 8, 96),
            nn.ReLU(),
            # Since there are four classification classes, the final layer must have 4 neurons
            nn.Linear(96, 4)
            # No softmax because cross entropy loss in PyTorch expects raw, unnormalized scores/logits and automatically applies log_softmax
        )

    def forward(self, x):
        return self.layers(x)

# For calculating mean loss and accuracy: New lists to store training and validation losses and accuracies for all trials
all_train_losses = []
all_val_losses = []
all_train_accuracies = []
all_val_accuracies = []    
    
# Training and evaluation
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=1234)
accuracies = []
losses = []

# 5 Trials
for train_idx, test_idx in kfold.split(x, y):
    val_idx = test_idx[:len(test_idx)//2]
    test_idx = test_idx[len(test_idx)//2:]

    train_dataset = TumorDataset(x[train_idx], y[train_idx])
    val_dataset = TumorDataset(x[val_idx], y[val_idx])
    test_dataset = TumorDataset(x[test_idx], y[test_idx])

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

    # Instantiate a new model
    model = TumorCNN()
    # Set loss function
    criterion = nn.CrossEntropyLoss()
    # Set optimizer
    optimizer = optim.Adam(model.parameters(), eps=0.01)

    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []
    
    # Batch loop (training epoch)
    for epoch in range(20):
        # Set to training mode
        model.train()
        train_loss = 0
        train_correct = 0
        for images, labels in train_loader:
            # Clears previous gradients to prevent accumulation
            # Ensures each batch's gradient update is based solely on current batch and prevents interference from previous batches
            optimizer.zero_grad()
            # Forward pass to calculate predictions
            outputs = model(images)
            # Calculate loss
            loss = criterion(outputs, labels)
            # Compute gradients of the loss wrt to each weight to determine how much to adjust each weight to minimize loss
            # read up on backpropagation
            loss.backward()
            # Update model weights
            optimizer.step()
            
            # To help calculate the average loss and accuracy per epoch
            train_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            train_correct += (predicted == labels).sum().item()
        
        # The training loss for each epoch is the average loss of all the images in the epoch batch
        train_loss /= len(train_loader)
        train_accuracy = train_correct / len(train_dataset)
        train_losses.append(train_loss)
        train_accuracies.append(train_accuracy)

        # Set to evaluation mode
        model.eval()
        val_loss = 0
        val_correct = 0
        # Disable gradient calculation for efficiency (not needed since we aren't changing the weights)
        with torch.no_grad():
            for images, labels in val_loader:
                # Predictions
                outputs = model(images)
                # Calculate loss
                loss = criterion(outputs, labels)
                # Calculate the average loss and accuracy
                val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                val_correct += (predicted == labels).sum().item()
        
        val_loss /= len(val_loader)
        val_accuracy = val_correct / len(val_dataset)
        val_losses.append(val_loss)
        val_accuracies.append(val_accuracy)
        print(f'Epoch {epoch+1}, Train Loss: {train_loss}, Train Accuracy: {train_accuracy}, Val Loss: {val_loss}, Val Accuracy: {val_accuracy}')
    
    # After each trial, store the losses and accuracies for averaging later
    all_train_losses.append(train_losses)
    all_val_losses.append(val_losses)
    all_train_accuracies.append(train_accuracies)
    all_val_accuracies.append(val_accuracies)
    
    # Test the final model accuracy
    model.eval()
    test_loss = 0
    test_correct = 0
    with torch.no_grad():
        for images, labels in test_loader:
            outputs = model(images)
            loss = criterion(outputs, labels)
            test_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            test_correct += (predicted == labels).sum().item()

    test_loss /= len(test_loader)
    test_accuracy = test_correct / len(test_dataset)
    accuracies.append(test_accuracy)
    losses.append(test_loss)

    # Plot learning curves
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    axes[0].plot(train_losses, label='Train Loss')
    axes[0].plot(val_losses, label='Validation Loss')
    axes[0].set_title('Loss')
    axes[0].legend()

    axes[1].plot(train_accuracies, label='Train Accuracy')
    axes[1].plot(val_accuracies, label='Validation Accuracy')
    axes[1].set_title('Accuracy')
    axes[1].legend()

    plt.show()

# Compute the mean of losses and accuracies across all trials
mean_train_losses = np.mean(all_train_losses, axis=0)
mean_val_losses = np.mean(all_val_losses, axis=0)
mean_train_accuracies = np.mean(all_train_accuracies, axis=0)
mean_val_accuracies = np.mean(all_val_accuracies, axis=0)

# Plot average learning curves across all trials
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(mean_train_losses, label='Mean Train Loss')
axes[0].plot(mean_val_losses, label='Mean Validation Loss')
axes[0].set_title('Mean Loss Across Trials')
axes[0].legend()

axes[1].plot(mean_train_accuracies, label='Mean Train Accuracy')
axes[1].plot(mean_val_accuracies, label='Mean Validation Accuracy')
axes[1].set_title('Mean Accuracy Across Trials')
axes[1].legend()

plt.show()
    
print('Accuracies:', accuracies)
print('Losses:', losses)
```

```python
# add losses and accurates to a dataframe

results = pd.DataFrame({
    'loss': losses,
    'accuracy': accuracies,
})

# Calculate the mean for 'loss' and 'accuracy'
mean_loss = results['loss'].mean()
mean_accuracy = results['accuracy'].mean()

# Calculate the standard deviation for 'loss' and 'accuracy'
std_loss = results['loss'].std()
std_accuracy = results['accuracy'].std()

# Append the mean and standard deviation values as new rows
results.loc['mean'] = [mean_loss, mean_accuracy]
results.loc['std'] = [std_loss, std_accuracy]


# print the mean
print("Mean Accuracy: {:.2f}".format(results['accuracy'].mean()))
print("Mean Loss: {:.2f}".format(results['loss'].mean()))
```
