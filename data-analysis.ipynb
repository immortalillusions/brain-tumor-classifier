```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def plot_best_fit(mean_values, std_dev, title):
    cases = [1, 2, 3, 4, 5]

    # Convert categorical x-axis labels to numerical values
    x_values = np.array(cases).reshape(-1, 1)
    y_values = np.array(mean_values)

    # Determine the best polynomial degree dynamically
    best_degree = 1
    best_mse = float('inf')
    best_model = None
    best_poly = None
    for degree in range(1, 6):  # Testing degrees 1 through 5
        poly = PolynomialFeatures(degree)
        X_poly = poly.fit_transform(x_values)
        model = LinearRegression().fit(X_poly, y_values)
        y_pred = model.predict(X_poly)
        mse = mean_squared_error(y_values, y_pred)
        if mse < best_mse:
            best_mse = mse
            best_degree = degree
            best_model = model
            best_poly = poly

    # Fit the best polynomial model
    X_poly = best_poly.fit_transform(x_values)
    curve_x = np.linspace(min(x_values), max(x_values), 100).reshape(-1, 1)
    curve_y = best_model.predict(best_poly.transform(curve_x))

    # Linear fit (Degree 1)
    linear_model = LinearRegression()
    linear_model.fit(x_values, y_values)
    linear_pred = linear_model.predict(x_values)

    # Print best-fit equation
    coefficients = best_model.coef_
    intercept = best_model.intercept_
    intercept += coefficients[0]
    coefficients = coefficients[1:]  # Remove the last coefficient
    poly_terms = [f"{coeff:.4f}x^{i+1}" for i, coeff in enumerate(coefficients)]
    equation = " + ".join(poly_terms[::-1])
    print(f"Best-fit polynomial degree: {best_degree}")
    print(f"Equation: y = {equation} + {intercept:.4f}")

    # Print linear fit equation
    linear_coefficients = linear_model.coef_
    linear_intercept = linear_model.intercept_
    print(f"Linear fit equation: y = {linear_coefficients[0]:.4f}x + {linear_intercept:.4f}")

    # Plot error bars
    plt.errorbar(x_values, mean_values, yerr=std_dev, fmt='o', label="Mean  Std Dev", capsize=5)
    plt.plot(curve_x, curve_y, 'r--', label=f"Best Fit: {equation} + {intercept:.4f}")
    plt.plot(x_values, linear_pred, 'b-', label=f"Linear Fit: y = {linear_coefficients[0]:.4f}x + {linear_intercept:.4f}")

    # Graph aesthetics
    plt.xticks(x_values.flatten(), cases)
    plt.xlabel("MaxPool2D Layers in Base")
    plt.ylabel("Mean Accuracy")
    plt.title(title)  # Using the provided title
    # Adjusting the legend
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=1, columnspacing=1.5, handlelength=2)
    plt.show()

# Calling
mean_values = [0.881724, 0.920357, 0.921842, 0.906092, 0.915899]
std_dev = [0.019046, 0.009549, 0.007905, 0.016526, 0.011050]
title = "Parameter 1: Architectures 1-5"

plot_best_fit(mean_values, std_dev, title)
```
